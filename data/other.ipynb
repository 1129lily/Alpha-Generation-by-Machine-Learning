# close price

```python
df_close = pd.read_csv("other/CRSP close price.csv")
# CRSP / Annual Update / Stock - Version 2 (CIZ) /Daily Stock File
df_close.query("Ticker in @ls_ticker",inplace=True)
```

# high/low price

```python
df_hl = pd.read_csv("other/CRSP high&low price.csv")
# CRSP / Annual Update / Stock - Version 2 (CIZ) /Daily Stock File
df_hl.query("Ticker in @ls_ticker",inplace=True)
```

```python
df_high = df_hl[['DlyCalDt','Ticker','DlyHigh']]
df_low = df_hl[['DlyCalDt','Ticker','DlyLow']]
```

```python
df_high1 = pd.pivot_table(df_high,index='DlyCalDt',columns='Ticker',values='DlyHigh')
df_low1 = pd.pivot_table(df_low,index='DlyCalDt',columns='Ticker',values='DlyLow')
```

```python
df_high1.interpolate(method='linear',inplace=True)
df_high1 = df_high1.bfill(axis=0)
df_low1.interpolate(method='linear',inplace=True)
df_low1 = df_low1.bfill(axis=0)
```

```python
df_high1.to_pickle("other/high price.pkl.zip",compression='zip')
df_low1.to_pickle("other/low price.pkl.zip",compression='zip')
```

# Rolling OLS prepartion

```python
df_return_list = pd.read_csv("return list.csv")
df_return_rolling = pd.read_csv("other/CRSP return(for rolling OLS).csv")
```

```python
ls_ticker_ols = list(df_return_list['Ticker'].unique())
```

```python
df_return_rolling.query("TICKER in @ls_ticker_ols",inplace=True)
df_return_rolling['date'] = pd.to_datetime(df_return_rolling['date'])
df_return_rolling.drop(['PERMNO','HEXCD'],axis = 1,inplace=True)
```

```python
df_return_rolling['RET'] = df_return_rolling['RET'].astype("float")
df_return_rolling.query("RET > -70 ",inplace=True)
df_return_rolling1 = pd.pivot_table(df_return_rolling, index = 'date', columns = 'TICKER', values = 'RET') *100
```

```python
df_null = df_return_rolling1.isnull().sum().to_frame()
df_null.columns = ['missing value']
df_null.reset_index(inplace = True)
```

```python
df_null.query("`missing value` > 0",inplace = True)

dic_null = dict(zip(df_null['TICKER'],df_null['missing value']))
for key, value in dic_null.items():
    if df_return_rolling1[key][:value+1].isnull().sum() != 0:
        print(key,df_return_rolling1[key][:value+1].isnull().sum())
```

```python
df_return_rolling2 = df_return_rolling1.interpolate(method='linear')
```

```python
df_null = df_return_rolling2.isnull().sum().to_frame()
df_null.columns = ['missing value']
df_null.reset_index(inplace = True)
```

```python
df_null.query("`missing value` > 0",inplace = True)
```

```python
df_return_rolling2.to_pickle("other/return(for rolling OLS).pkl.zip",compression='zip')
```
