```python
import numpy as np
import pandas as pd
import os
```

# Stock List required

```python
df_info = pd.read_csv('other/Stock List.csv')
ls_PERMNO = list(df_info['PERMNO'])
dict_PERMNO = dict(zip(ls_PERMNO, df_info['HTICK']))
```

# Concat all the factor files

```python
df_factor_CRSP = pd.DataFrame()
for file in os.listdir("./factor/CRSP/"):
    # Factors by WRDS / Factors (Backtester) by WRDS
    if '0' not in file:
        df1 = pd.read_csv("./factor/CRSP/" + file)
        df1.query('PERMNO in @ls_PERMNO',inplace=True)
        df1 = df1[df1.columns[:3]]
        if len(df_factor_CRSP)>0:
            df_factor_CRSP = pd.merge(df_factor_CRSP,df1,how='outer',on=['fdate','PERMNO'])
        else:
            df_factor_CRSP = df1
        print(file+" already merged")
```

```python
df_factor_CRSP['Ticker'] = df_factor_CRSP['PERMNO'].map(dict_PERMNO)
df_factor_CRSP['Date'] = pd.to_datetime(df_factor_CRSP['fdate'])
```

# Stock return

```python
df_return_long = pd.read_pickle("return/stock return(long).pkl.zip")
df_return_long['date'] = pd.to_datetime(df_return_long['date'])
df_return_long['year'] = df_return_long['date'].dt.year
df_return_long['month'] = df_return_long['date'].dt.month
```

```python
df_factor_CRSP['Date_match'] = df_factor_CRSP['Date'] + pd.Timedelta(days=1)
df_factor_CRSP['year'] = df_factor_CRSP['Date_match'].dt.year
df_factor_CRSP['month'] = df_factor_CRSP['Date_match'].dt.month
```

# Return with Factor

```python
df_factor_matched = pd.merge(df_return_long,df_factor_CRSP,how='left',left_on=['year','month','Ticker'],right_on=['year','month','Ticker'])
df_factor_matched = df_factor_matched[['date','Ticker','stock return']+ls_variablename]
```
