```python
# df_ff.drop(['Mkt-RF','RF'], axis=1, inplace=True)
df_alpha['date'] = pd.to_datetime(df_alpha['date'])
df_ff['date'] = pd.to_datetime(df_ff['date'])
df_return['date'] = pd.to_datetime(df_return['date'])
df_return_list['date'] = pd.to_datetime(df_return_list['date'])
df_MA['date'] = pd.to_datetime(df_MA['date'])
df_momentum['date'] = pd.to_datetime(df_momentum['date'])
df_vol['date'] = pd.to_datetime(df_vol['date'])
df_volume['date'] = pd.to_datetime(df_volume['date'])
df_overlap['date'] = pd.to_datetime(df_overlap['date'])
df_CRSP['date'] = pd.to_datetime(df_CRSP['date'])
```
